**Word embeddings** are machine learning techniques that represent words as numerical vectors. These vector representations capture semantic and contextual relationships between words and are used in many applications in natural language processing (NLP). Word embeddings allow measuring the similarities between words and developing language models by trying to express the meanings of words in a mathematical form.

Common Uses: Word2Vec and GloVe

1. Word2Vec (Word to Vector):
   - Developed by Google.
   - There are two main approaches to converting words into vectors: Continuous Bag of Words (CBOW) and Skip-gram.
     - CBOW: Predicts the target word using surrounding words.
     - Skip-gram: Predicts surrounding words using the target word.
   - The purpose of the model is to learn the meanings of words according to their context. In this way, similar words are represented by similar vectors.

2. GloVe (Global Vectors for Word Representation):
   - Developed by Stanford.
   - Creates word vectors using the probability of words occurring together.
   - GloVe relies on global contextual information between word pairs and evaluates word similarities in a global context.

**Word2Vec Model with Gensim**

```python
from gensim.models import Word2Vec
import warnings
warnings.filterwarnings("ignore")
```

```python
sentences = [
  "Natural language processing allows machines to understand and interpret human language.",
  "Artificial intelligence is used to build models that can predict customer behavior."
]
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
tokens = [word_tokenize(sentence) for sentence in sentences]
print(tokens)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
text = []
for i in sentences:
    text.append(i)
text = ''.join(map(str, text))
wordcloud = WordCloud(width=6000, height=600, max_font_size=300,background_color='white').generate(text)
plt.figure(figsize=(20,17))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# Word2Vec model creating
model = Word2Vec(tokens, vector_size=100, window=5, min_count=1, workers=4)
# vector_size: How many dimensional vector each word will be represented by.
# window: Determines how wide the context of words will be.
# min_count: Sets how frequently occurring words will be taken into account.
# workers: Number of processor cores to be used during training.
```

```python
# Finding similar words of a word from the model
# This code returns the three words closest to human.
similar_words=model.wv.most_similar('human',topn=3)
print(similar_words)
```

```python
# measuring similarity between two words
similarity=model.wv.similarity('Natural','Artificial')
print(f"Similarity (Natural vs Artificial): {similarity}")
```
